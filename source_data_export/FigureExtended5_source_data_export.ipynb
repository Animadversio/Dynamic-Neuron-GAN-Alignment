# Extended Figure 5 — Source Data Export

**Extended Figure 5** investigates latent code linearity and the properties
of RealNVP-based GAN priors.  Panel 5C compares Adam and CMA-ES optimization
performance as a function of RealNVP layer depth and latent dimensionality.
Panels 5D/E show R² values measuring how well linear interpolations in
BigGAN and DeePSim latent spaces predict neural responses (Caffenet scorer).

> `insilico_experiments/latent_code_linearity.py` and are not distributed

The exported source data files are written to `../source_data/ExtendedFig5/`.

````python
import os, sys, pickle, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from os.path import join

sys.path.insert(0, os.path.join(os.path.dirname(os.getcwd()), ".."))
from neuro_data_analysis.neural_data_lib import (
    load_neural_data,
    extract_all_evol_trajectory_psth,
    pad_psth_traj,
    get_expstr,
    extract_natref_activation_array,
    extract_evol_activation_array,
)
from neuro_data_analysis.neural_data_utils import get_all_masks
````

````python
# Set the path to pre-computed in-silico experiment results.
# Override by setting the MATROOT environment variable before launching Jupyter.
matroot = os.environ.get("MATROOT", "")

# Directory containing RealNVP optimization and linearity results
realnvpdir = join(matroot, "RealNVP_Results")
outdir = "../source_data/ExtendedFig5"
os.makedirs(outdir, exist_ok=True)
````

## Extended Figure 5C: RealNVP optimization

Load and export the Adam and CMA-ES optimization score tables as a function
of RealNVP layer depth and latent dimensionality (log-scale, quadratic
fitness landscape).

````python
# Adam optimization results: rows = depth levels, cols = latent dims
adam_pkl = pickle.load(open(
    join(realnvpdir, "Adam_optim_score_as_function_of_depth_and_dim_logscale_quad_case.pkl"), "rb"
))
adam_df = pd.DataFrame(adam_pkl) if not isinstance(adam_pkl, pd.DataFrame) else adam_pkl
adam_df.to_csv(
    join(outdir, "FigureExt5C_Adam_optim_score_as_function_of_depth_and_dim_logscale_quad_case.csv"),
    index=False
)
pickle.dump(adam_df, open(
    join(outdir, "FigureExt5C_Adam_optim_score_as_function_of_depth_and_dim_logscale_quad_case.pkl"), "wb"
))

# CMA-ES optimization results
cmaes_pkl = pickle.load(open(
    join(realnvpdir, "realnvp_layer_dim_cmaes_optim_result_logscale_quad_case.pkl"), "rb"
))
cmaes_df = pd.DataFrame(cmaes_pkl) if not isinstance(cmaes_pkl, pd.DataFrame) else cmaes_pkl
cmaes_df.to_csv(
    join(outdir, "FigureExt5C_realnvp_layer_dim_cmaes_optim_result_logscale_quad_case.csv"),
    index=False
)
pickle.dump(cmaes_df, open(
    join(outdir, "FigureExt5C_realnvp_layer_dim_cmaes_optim_result_logscale_quad_case.pkl"), "wb"
))

print(f"Adam result shape:  {adam_df.shape}")
print(f"CMA-ES result shape: {cmaes_df.shape}")
print("Extended Figure 5C source data saved.")
````

## Extended Figure 5D/E: Latent code linearity

Load and export the R² tables measuring how well linear interpolations
in BigGAN and DeePSim latent spaces predict Caffenet unit responses.

````python
lineardir = join(matroot, "Latent_Linearity")

# BigGAN Caffenet R² results (Extended Figure 5D)
biggan_lin_df = pd.read_csv(join(lineardir, "BigGAN_Caffenet_latent_code_linearity_results.csv"))
biggan_lin_df.to_csv(
    join(outdir, "FigureExt6D_BigGAN_Caffenet_latent_code_linearity_results.csv"), index=False
)

# DeePSim FC6 Caffenet R² results (Extended Figure 5E)
deepsim_lin_df = pd.read_csv(join(lineardir, "DeePSim_FC6_Caffenet_latent_code_linearity_results.csv"))
deepsim_lin_df.to_csv(
    join(outdir, "FigureExt6E_DeePSim_FC6_Caffenet_latent_code_linearity_results.csv"), index=False
)

print(f"BigGAN linearity df shape:   {biggan_lin_df.shape}")
print(f"DeePSim linearity df shape:  {deepsim_lin_df.shape}")
print("Extended Figure 5D/E source data saved to", outdir)
````
