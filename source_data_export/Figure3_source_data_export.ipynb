## Figure 3G: Image distance vs PSTH similarity

Load the pre-computed image distance dataframe and the activation meta

````python
outdir = "../source_data/Fig3"
os.makedirs(outdir, exist_ok=True)

# Load pre-computed dataframes from the analysis pipeline
imgdist_df = pd.read_csv(join(tabdir, "proto_imdist_psth_covstr_sim_df.csv"))
meta_act_df = pd.read_csv(join(tabdir, "meta_act_df.csv"))

# Export
imgdist_df.to_csv(join(outdir, "Fig3G_proto_imdist_psth_covstr_sim_df.csv"), index=False)
meta_act_df.to_csv(join(outdir, "Fig3_meta_df.csv"), index=False)

print(f"imgdist_df shape: {imgdist_df.shape}")
print(f"meta_act_df shape: {meta_act_df.shape}")
print("Figure 3 source data saved to", outdir)
````
